```python
BUS = 'Megabus (bus)'
DATESCRAPED = '2016-06-20'
ORIGINCITY = 'Washington, DC'
```

```python
sf = gl.SFrame.from_odbc(db, """ SELECT TO_CHAR(origintime, 'HH24:MI:SS') as origintime2,
                                    TO_CHAR(destinationtime, 'HH24:MI:SS') as destinationtime2,
                                    * FROM trips_prod 
                                    WHERE (carrier ='{}'
                                    AND CAST(datescraped AS DATE) = '{}'
                                    AND origincity = '{}')
                                    """.format(BUS, DATESCRAPED, ORIGINCITY))
```

```python
sf.show(view = 'Scatter Plot', x = 'dateoftrip', y='price')
```

```python
time = sf.groupby(key_columns='origintime2',
                               operations={
                                    'mean_price': agg.MEAN('price'),
                                    'std_price': agg.STD('price'),
                                    'count': agg.COUNT('tripid')
                                })
time.sort('origintime2').print_rows(num_rows=39, num_columns=4)
```

```python
date = sf.groupby(key_columns='dateoftrip',
                               operations={
                                    'mean_price': agg.MEAN('price'),
                                    'std_price': agg.STD('price'),
                                    'count': agg.COUNT('tripid')
                                })
date.sort('dateoftrip').print_rows(num_rows=39, num_columns=4)
```
